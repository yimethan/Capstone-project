```python
all_df = pd.concat([df, pd.read_excel(os.path.join(path, 'uic_expo.xlsx'))])
```

```python
all_df.loc[all_df['result'] == '봄 웜톤', 'result'] = 'spring'
```

```python
all_df.loc[all_df['result'] == '가을 웜톤', 'result'] = 'fall'
```

```python
all_df.loc[all_df['result'] == '여름 쿨톤', 'result'] = 'summer'
```

```python
all_df.loc[all_df['result'] == '겨울 쿨톤', 'result'] = 'winter'
```

```python
all_df.result.unique()
```

```python
all_df.result.value_counts().plot.bar()
```

```python
seasons = ['spring', 'summer', 'fall', 'winter']
for season in seasons:
    print(len(all_df[all_df['result'] == season]))
```

```python
import matplotlib.pyplot as plt
```

```python
for season in seasons:
    plt.figure()
    all_df[all_df['result'] == season].light_num.value_counts().plot.bar()
    plt.savefig(f'/content/graph/{season}_light.png')
    all_df[all_df['result'] == season].frame_num.value_counts().plot.bar()
    plt.savefig(f'/content/graph/{season}_frame.png')
```
